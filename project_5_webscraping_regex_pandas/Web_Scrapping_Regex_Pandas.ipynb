Web Scrapping + Regular Expression + Pandas

```python
#finding the most used words dfrom a transcript

from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
soup.find_all('p')
```

```python
mlkj_speech = soup.find_all('p')

type(mlkj_speech)
```

```python
speech_combined = [p.text for p in mlkj_speech]
print(speech_combined)
```

```python
type(speech_combined)
```

```python
' '.join(speech_combined)

string_speech = ' '.join(speech_combined)
```

```python
# Replace carriage returns/newlines with a space
string_speech_cleaned = string_speech.replace('\r\n', ' ')

print(string_speech_cleaned)
```

```python
import re
```

```python
string_speech_cleaned = string_speech.replace('\r\n', ' ')
result = re.sub(r'[^\w\s]', '', string_speech_cleaned)
```

```python
print(result)
```

```python
speech_lower = result.lower()
re.split(r'\s+', speech_lower)
```

```python
speech_broken_out = re.split(r'\s+', speech_lower)
```

```python
import pandas as pd
```

```python
# Create DataFrame with one column named 'word'
df = pd.DataFrame(speech_broken_out, columns=['word'])

# Count the words
word_counts = df['word'].value_counts()

print(word_counts.head(20))  # top 20 most used words
```
